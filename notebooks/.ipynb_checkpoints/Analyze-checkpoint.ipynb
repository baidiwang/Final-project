```python
import pandas as pd
import numpy as np
```

```python
#Analyze states
df_states = pd.read_csv('../data/Immigrants States with most cases(by March2020).csv')
```

```python
df_states.head()
```

```python
df_states.tail()
```

```python
df_states.isna().sum()
```

```python
df_states = df_states.dropna()
df_states
```

```python
df_states = df_states.rename(columns={'All': 'States'})
```

```python
df_states = df_states.rename(columns={'4,830,173': 'Deportation Cases'})
df_states
```

```python
# The forth colume is "Not Known" so I have to drop it
df_states.drop([4])
```

```python
#I have to remove commas in Deportation Cases Colume, as numbers with commas won't work in Datawrapper
df_states['Deportation Cases'] = df_states['Deportation Cases'].str.replace(',', '')
df_states
```

```python
df_states.to_csv('../output/by_states.csv', index=False)
```
